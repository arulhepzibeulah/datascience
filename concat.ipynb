```python
#Concatenation
import  pandas as pd
df1=pd.DataFrame({'a':['hepzi','bobby','chalk'],'b':['aa','bb','cc']})
df2=pd.DataFrame({'a':['whaya','nigga','cccoc'],'b':['aasda','bbfs','ccas']})
```

```python
df_concat=pd.concat([df1,df2 ], ignore_index=True)
df_concat
```

```python
df_concat_cols = pd.concat([df1, df2], axis=1)
df_concat_cols
```

```python
one=pd.DataFrame({
    'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5'],
   'Marks_scored':[98,90,87,69,78]})

two = pd.DataFrame({
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5'],
   'Marks_scored':[89,80,79,97,88]})
 
```

```python
one,two
```

```python
pd.merge(one,two,on='subject_id',how='inner')
```

```python
pd.merge(one,two, on='subject_id',how='left')
```

```python
pd.merge(one,two, on='subject_id',how='right')
```

```python
pd.merge(one,two, on='subject_id',how='outer')
```
